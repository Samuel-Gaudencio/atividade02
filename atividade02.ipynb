```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = '/content/imagem.jpg'
imagem = cv2.imread(image_path)
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
```

```python
print(f'Altura: {imagem.shape[0]} \nLargura: {imagem.shape[1]}\nCanais de cor: {imagem.shape[2]}\nTipo de dado: {imagem.dtype} \nTom de cinza máximo: {imagem.max()} \nTom de cinza médio: {imagem.mean()} \nTom de cinza mínimo: {imagem.min()}')
```


2. Faça a leitura de uma imagem de entrada 2D colorida qualquer e aplique o
filtro da média, mediana e gaussiano na imagem de entrada. Mostre os
resultados de cada filtro por meio da biblioteca matplotlib



```python
blurred_image = cv2.blur(imagem, (5, 5))
plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Imagem sem filtro')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Imagem com filtro da média')
```

```python
blurred_image = cv2.medianBlur(imagem, 5)
plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Imagem sem filtro')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Imagem com filtro Mediana')
```

```python
blurred_image = cv2.GaussianBlur(imagem, (7, 7), 0)
plt.subplot(1, 2, 1)
plt.imshow(imagem_rgb)
plt.title('Imagem sem filtro')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Imagem com filtro Gaussian')
```
